# Importing Numpy , Pandas and Matplotlib 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Taking values

```python
x = [0,2,4,6,8]
y = [0,4,16,36,64]
```

```python
fig,ax = plt.subplots()
ax.plot(x,y,marker='o',label="Data Points")
ax.set_title("Basic Components of Matplotlib Figure")
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")
ax.legend()
plt.show()
```

```python
arr = np.array([10,20,30,40])
arr2 = np.array([50,25,90,60])
df=pd.DataFrame({"data_1":arr,"data_2":arr2})

fig,ax = plt.subplots()
ax.plot(df.data_1,df.data_2,marker = 'o',label="DataPoints")
ax.legend()
plt.show()
```

```python
plt.bar(df.data_1,df.data_2,width=5)
plt.show()
```
