# Histogram : visualization of frequency

df.Make.value_counts(): Counts the occurrences of each unique value in the Make column.

.nlargest(40): Selects the top 40 values with the highest counts.

.plot(kind='bar', figsize=(10, 5)): Plots these values as a bar chart with a figure size of 10x5

```python
df.Make.value_counts().nlargest(40).plot(kind='bar',figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel("Number of cars")
plt.xlabel('Make')
```

# Data Transformation: Applying Custom Functions

```python
def get_year(year):
    if year<2011:
        return 'old'
    elif 2011<=year<2018:
        return'Medium'
    else:
        return 'new'
    
df['type']=df['Year'].apply(get_year)

df.head(2)
    
```
